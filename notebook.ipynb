<h2 style='font-family: Colonna MT; font-weight: 600; font-size: 20px; text-align: left'>2.0. Generating Same Invoice on Multiple Clients</h2>
  

This system accommodates diverse data storage formats through flexible preprocessing bridges, ensuring consistent invoice generation regardless of source structure. When handling **columnar JSON** (spreadsheet-like nested dictionaries), we transform the data into a list of client dictionaries for standardized processing. For **CSV/Excel files**, pandas efficiently loads and converts tabular data into the same uniform format. With **simple Python dictionaries**, we leverage direct key-value access when data is already properly nested, or implement lightweight parsing for flat dictionary structures. The architecture intentionally isolates the data ingestion layer from the invoice generation logic - once converted to our standardized list-of-dictionaries format, all subsequent processing remains identical. This design allows teams to maintain their preferred data storage method (whether JSON documents, spreadsheets, or database exports) while guaranteeing reliable PDF output. The system's adaptability stems from this crucial normalization step, which adds minimal overhead while providing maximum compatibility. Future extensions can incorporate additional data bridges (like SQL queries or API connectors) without modifying the core invoice rendering pipeline, making the solution both robust to current requirements and adaptable to future data source variations.

<h3 style='font-family: Colonna MT; font-weight: 600; font-size: 20px; text-align: left'>B. Option 2: Data store in JSON</h3>

- This system automates multi-client invoice generation by restructuring columnar JSON data (where each field contains indexed client entries) into a list of cohesive client dictionaries—each containing name, contact details, and address. By iterating through this normalized dataset, it produces customized PDF invoices with predefined items, calculated totals, and client-specific branding, saving outputs with unique filenames for traceability. The approach ensures scalability (handling hundreds of clients), maintains consistency, and adapts to varying input structures while separating data processing from PDF logic for clean maintenance. Future enhancements could include dynamic invoice numbering or automated delivery integrations.

```python
from Invoice.Invoicepdf import PDFInvoice
import json

# Load client info from JSON file
with open("Datasets/clients.json", "r") as file:
    clients_data = json.load(file)

# Transform the dictionary of dictionaries into a list of client dictionaries
clients = []
for i in range(len(clients_data['customer_name'])):
    client = {
        'customer_name': clients_data['customer_name'][str(i)],
        'customer_email': clients_data['customer_email'][str(i)],
        'customer_phone': clients_data['customer_phone'][str(i)],
        'customer_address': clients_data['customer_address'][str(i)]
    }
    clients.append(client)

# Generate invoices for each client
for client in clients:
    pdf = PDFInvoice(
        company_name="Mizer Companition",
        tagline="Creative Business Idea",
        invoice_number="INV-20250509",
        invoice_date="May 9, 2025",
        terms="Payment due in 30 days.\nNo refunds after 7 days.",
        payment_info="Account: Mizer Inc\nBank: GlobalBank\nIBAN: GB00 1234 5678 9012"
    )

    pdf.set_margins()
    pdf.add_page()
    pdf.add_invoice_info(
        to_info=[
            client["customer_name"],
            f"Address: {client['customer_address']}",
            f"Phone: {client['customer_phone']}",
            f"Email: {client['customer_email']}"
        ],
        from_info=[
            "Jane Doe",
            "Managing Director, Company Ltd",
            "Phone: +123 4567 8910",
            "Email: janedoe@gmail.com"
        ]
    )

    subtotal = pdf.add_table(items)
    pdf.add_summary_section(subtotal=subtotal, tax_rate=0.15, discount=100)
    filename = f"Invoice_{client['customer_name'].replace(' ', '_')}"
    pdf.output(f"Outputs/{filename}.pdf")
    log_success(filename)
```

<h3 style='font-family: Colonna MT; font-weight: 600; font-size: 20px; text-align: left'>A. Option 2: Pandas-based approach</h3>


For developers already working with pandas or handling large datasets, an alternative method restructures the raw JSON data using DataFrame operations for cleaner transformation. By loading the columnar JSON (where each key maps to client attributes) into a pandas DataFrame, we leverage its built-in `to_dict('records')` method to automatically convert rows into a list of client dictionaries—each containing all relevant fields (name, email, phone, etc.) as key-value pairs. This approach is particularly efficient for bulk processing, as pandas handles the iteration and restructuring implicitly, reducing manual data wrangling. Once transformed, the standardized list of dictionaries integrates seamlessly with the existing invoice generation logic, ensuring consistency while benefiting from pandas’ performance optimizations for large datasets. This method is ideal for teams already embedded in the Python data ecosystem or when additional preprocessing (e.g., filtering missing values, merging external data) is required before invoice creation.

```python
from Invoice.Invoicepdf import PDFInvoice
from Utilis.logger import log_success
from datetime import datetime
import pandas as pd
import json
import os

# Load client info from JSON file
with open("Datasets/clients.json", "r") as file:
    clients_data = json.load(file)
    

# Convert the dictionary to a pandas DataFrame
df = pd.DataFrame(clients_data)

# Convert the DataFrame to a list of dictionaries
clients = df.to_dict('records')

# Generate invoices for each client
for client in clients:
    pdf = PDFInvoice(
        company_name="Mizer Companition",
        tagline="Creative Business Idea",
        invoice_number="INV-20250509",
        invoice_date="May 9, 2025",
        terms="Payment due in 30 days.\nNo refunds after 7 days.",
        payment_info="Account: Mizer Inc\nBank: GlobalBank\nIBAN: GB00 1234 5678 9012"
    )

    pdf.set_margins()
    pdf.add_page()
    pdf.add_invoice_info(
        to_info=[
            client["customer_name"],
            f"Address: {client['customer_address']}",
            f"Phone: {client['customer_phone']}",
            f"Email: {client['customer_email']}"
        ],
        from_info=[
            "Jane Doe",
            "Managing Director, Company Ltd",
            "Phone: +123 4567 8910",
            "Email: janedoe@gmail.com"
        ]
    )

    subtotal = pdf.add_table(items)
    pdf.add_summary_section(subtotal=subtotal, tax_rate=0.15, discount=100)
    filename = f"Invoice_{client['customer_name'].replace(' ', '_')}"
    pdf.output(f"Outputs/{filename}.pdf")
    log_success(filename)
```

<h2 style='font-family: Colonna MT; font-weight: 600; font-size: 20px; text-align: left'>2.0. Generating Unique Invoices for Multiple Clients</h2>

```python
from Invoice.Invoicepdf import PDFInvoice
from Utilis.logger import log_success
from datetime import datetime
import pandas as pd
import os


df = pd.read_csv("Datasets/orders.csv")

# Group data by cart_id  as Its Unique Id (each cart_id = one invoice)
for cart_id, group in df.groupby("cart_id"):
    customer = group.iloc[0]

    # Prepare recipient (customer) information
    to_info = [
        customer["customer_name"],
        f"Phone: {customer['customer_phone']}",
        f"Email: {customer['customer_email']}",
        f"Addres: {customer['customer_address']}"
    ]

    # Static sender (company) information
    from_info = [
        "Jane Doe",
        "Managing Director, Company Ltd",
        "Phone: +123 4567 8910",
        "Email: janedoe@gmail.com"
    ]

    # Prepare list of purchased items
    items = []
    for _, row in group.iterrows():
        items.append({
            "description": row["product_description"],
            "unit_price": float(row["unit_price"]),
            "quantity": int(row["quantity"])
        })

    # Get tax rate and total discount
    tax_rate = float(customer["tax_rate"])
    discount = float(group["discount"].sum())

    # Create PDF invoice
    pdf = PDFInvoice(
        company_name="Mizer Companition",
        tagline="Creative Business Idea",
        invoice_number=cart_id,
        invoice_date=datetime.today().strftime('%B %d, %Y'),
        terms="Payment due in 30 days.\nNo refunds after 7 days.",
        payment_info="Account: Mizer Inc\nBank: GlobalBank\nIBAN: GB00 1234 5678 9012"
    )

    # Build the invoice layout
    pdf.set_margins()
    pdf.add_page()
    pdf.add_invoice_info(to_info=to_info, from_info=from_info)
    subtotal = pdf.add_table(items)
    pdf.add_summary_section(subtotal=subtotal, tax_rate=tax_rate, discount=discount)

    # Save the invoice as PDF
    os.makedirs("Outputs", exist_ok=True)
    pdf.output(f"Outputs/{cart_id}.pdf")

    # Log success
    log_success(cart_id)
```
